```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
girl = cv2.imread("image_17/1782372.jpg")
girl = cv2.cvtColor(girl, cv2.COLOR_BGR2RGB)
plt.imshow(girl)
```

```python
new_girl = np.zeros(girl.shape, girl.dtype)
```

```python
alpha = 1.1
beta = 0
```

```python
for y in range(girl.shape[0]):
    for x in range(girl.shape[1]):
        for c in range(girl.shape[2]):
            new_girl[y, x, c] = np.clip(alpha*girl[y, x, c] + beta, 0, 255)
```

```python
plt.imshow(new_girl)
```
